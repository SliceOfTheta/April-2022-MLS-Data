```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv("Orlando.csv")
df = df.rename(columns={'Property Type': "PT", 'Status': "Stat", 'Property Style': "PS", 'List Price': "LP", 'List Date': "LD",
       'City Name': "CN", 'Zip Code': "ZIP", 'Year Built': "YEAR", 'Heated (Living) Area': "SQFT",
       'Bedrooms': "BEDS", 'Bathrooms': "BATHS", 'Pool': "POOL", 'Waterfront (Y/N)': "WATER", 'Sale Price': "SP",
       'Sale Date': "SD"})
df.head(100)  # To display the top 100 rows 
```

```python
df.dtypes
print
```

```python
df.PS.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of Sales by Property Type")
plt.ylabel('Sales')
plt.xlabel('Property Style');
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['LP'], df['SP'])
ax.set_xlabel('List Price')
ax.set_ylabel('Sale Price')
plt.show()
```
